# Overview of the Dataset



This dataset contains information about the daily activities and academic performance of 2,000 students. The purpose of this data is to explore how various lifestyle factors — such as study habits, sleep duration, physical activity, and social interactions — relate to student stress levels and academic grades.
| Feature                           | Description                                                             |
| --------------------------------- | ----------------------------------------------------------------------- |
| `Student_ID`                      | A unique identifier for each student (not useful for prediction).       |
| `Study_Hours_Per_Day`             | Number of hours the student studies each day.                           |
| `Extracurricular_Hours_Per_Day`   | Hours spent on extracurricular activities like clubs, arts, or hobbies. |
| `Sleep_Hours_Per_Day`             | Average daily sleep duration (in hours).                                |
| `Social_Hours_Per_Day`            | Time spent daily in social interaction (with friends, family, etc.).    |
| `Physical_Activity_Hours_Per_Day` | Daily physical exercise or sports time.                                 |
| `Stress_Level`                    | Student's self-reported stress level (Low, Moderate, or High).          |
| `Gender`                          | Gender of the student (Male or Female).                                 |
| `Grades`                          | Final academic grade (numeric score, e.g., GPA or percentage).          |

# Data preprocessing

1-Drop unnecessary columns – The Student_ID column is not useful for prediction, so we remove it.

2-Encode categorical features:

- Stress_Level is an ordinal variable, so we map it manually:
        'Low' → 0, 'Moderate' → 1, 'High' → 2.

- Gender is a nominal variable, so we apply label encoding. This converts categories like "Male" and "Female" to numeric values (e.g., Male=1, Female=0).

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split

# Load the dataset
df = pd.read_csv("student_lifestyle_dataset..csv")

# Drop the 'Student_ID' column
df.drop(columns=['Student_ID'], inplace=True)

# Ordinal encoding for 'Stress_Level'
stress_mapping = {'Low': 0, 'Moderate': 1, 'High': 2}
df['Stress_Level'] = df['Stress_Level'].map(stress_mapping)

# Label encoding for 'Gender'
label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])  # Typically Male=1, Female=0
```

1. Feature Separation:

    - Divide Dataset into Features and Target.

2. Feature Scaling with StandardScaler:

    - Prevents domination by high-magnitude features.

    - Improves performance of many machine learning algorithms.

3. Train-Test Split Strategy:

    - Training Set Size: 80% of total data.

    - Random state ensures consistent splitting across runs.



```python
# Separate features and target
X = df.drop(columns=['Grades'])
y = df['Grades']

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
# Optional: Show shape of datasets
print("Training features shape:", X_train.shape)
print("Testing features shape:", X_test.shape)
print("Training labels shape:", y_train.shape)
print("Testing labels shape:", y_test.shape)
```

# Predictive modeling:

**In this section, we build and evaluate a model to predict student grades based on input features like stress level, gender, sleep hours, etc.**

1. Initialize the model.

2. Train the model using the training data (X_train, y_train).

3. Predict the grades for the test data (X_test).

4. Evaluate the model using:

    - Mean Squared Error (MSE): Measures average squared difference between actual and predicted grades. Lower is better.

    - R-squared (R²): Indicates how well the model explains the variability of the target variable. Ranges from 0 to 1 (higher is better).

## Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("R-squared (R²):", r2)
```

**Key insights:**
- The Mean Squared Error (MSE) is 0.265, indicating that on average, the squared difference between the actual and predicted grades is relatively low — but there is still room for improvement.

- The Linear Regression model achieved an R² score of 0.546, meaning it was able to explain approximately 54.6% of the variance in students' grades based on the given features. While this indicates a moderate predictive power, it also suggests that nearly half of the variation in grades is influenced by other factors not captured in the dataset.

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize Random Forest Regressor
rf_model = RandomForestRegressor(
    n_estimators=100,  
    random_state=42   
)

rf_model.fit(X_train, y_train)

y_rf_pred = rf_model.predict(X_test)

rf_mse = mean_squared_error(y_test, y_rf_pred)
rf_r2 = r2_score(y_test, y_rf_pred)

print("Random Forest MSE:", rf_mse)
print("Random Forest R²:", rf_r2)
```

**Key insights:**
1. An R² of 0.455 means the model explains about 45.5% of the variance in student grades.
2. This is a moderate result: better than guessing, but still far from perfect.
3. MSE of ~0.318 suggests that, on average, predictions deviate from actual grades by √0.318 ≈ 0.56 grade points.
4. Depending on the grading scale (e.g., out of 10), this could be acceptable or need tightening.
